```python
tables = df[df['Sub-Category'] == 'Tables']

state_margin = (
    tables.groupby('State')
    .agg({
        'Sales':'sum',
        'Profit':'sum'
    })
)

state_margin['Profit_Margin_%'] = (
    state_margin['Profit']
    / state_margin['Sales']
) * 100

state_margin.sort_values('Profit_Margin_%')
```
