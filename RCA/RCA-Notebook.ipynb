```python
df[df['Sub-Category'] == 'Tables'] \
.groupby('Discount') \
.agg({
    'Sales':'sum',
    'Profit':'sum',
    'Order ID':'count'
})
```

### Root Cause
    - Furniture losses are mainly caused by the Tables sub-category.
    - Tables become unprofitable once discounts exceed roughly 20%.
    - Discounts between 30%-50% generate severe losses.
    - States such as New York, Illinois, North Carolina, Ohio, and Tennessee are major contributors because many table sales there occur at these high discount levels.
    - Tables sold without discounts achieve an 18.55% profit margin, proving the product itself is profitable.
